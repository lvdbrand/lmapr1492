```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import display

cle_api="DFGNkaXUTedPM8g2"
id_mp="mp-1018096"
with MPRester(cle_api) as l:
    structure=l.get_structure_by_material_id(id_mp)
    
calculator=XRDCalculator()
pattern=calculator.get_pattern(structure)
hkls=pattern.hkls
angle=pattern.x
intensity=pattern.y
```

```python
print("Premier pic du diffractogramme :")
print("Indices de Miller associés : ",hkls[0][0]['hkl'])
print("Multiplicité des indices de Miller associés :",hkls[0][0]['multiplicity'])
print("Angle : 2\u03B8 =", angle[0])
print("Intensité :",intensity[0])
```
