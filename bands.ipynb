```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen import Lattice, Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.electronic_structure import plotter
from pymatgen.electronic_structure import bandstructure as bs
from pymatgen.electronic_structure.plotter import BSPlotter
from pymatgen import Structure
from matplotlib import pyplot as plt

cle_api="DFGNkaXUTedPM8g2"
id_mp="mp-1018096"
with MPRester(cle_api) as l:
    struct=l.get_structure_by_material_id(id_mp)
    bands=l.get_bandstructure_by_material_id(id_mp)
```

```python
print("Structure de la bande :")
plotter.BSPlotter(bands).show()
```

```python
print("1. Dterminer la bande interdite \n")
band_gap=bands.get_band_gap()
print("La bande interdite est-elle direct? :",band_gap['direct'])
print("Energie de la bande interdite : ", band_gap['energy'], "[eV]")
print("Transition :",band_gap['transition']+'\n')
print("Energie de Fermi :", bands.efermi,"[eV]\n")
print("Puisque l'nerge de la bande interdite est suprieure  1eV, on peut conclure que le Ba2NF est un isolant.")
```

```python
print("2. Trouver la direction dans laquelle la dernire (premire) bande de valence (conduction) prsente le plus et le moins de dispersion.\n")
print("La dispersion est la plus importante (la plus faible) lorsque la pente de l'nergie en fonction du vecteur d'onde est la plus haute (la plus basse).\n")
```
